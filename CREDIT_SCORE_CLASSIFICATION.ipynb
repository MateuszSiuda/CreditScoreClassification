## Introduction
<ins>Credit Score</ins> determines the probability that the borrower will pay back the loan on time. <br>

To define client scoring companies use many information which the most important are:
- payment history,
- current unpaid debt,
- the number and type of loan accounts,
- lenght of period of time that loan accounts is open,
- usage of available credit,
- the number of credit inquiries,

This is very important to have a good credit score because it can help us get a loan for a new IPhone, car, house but also get a new policy or rent an apartment.

**In this project we will try to classify client credit score into three buckets: Poor, Standard, Good, having available person's credit-related information**

Dataset come from Kaggle where we can find short description about project and data:<br>
https://www.kaggle.com/datasets/parisrohan/credit-score-classification

### Removing errors and applying assumption
<br> 
To applay rules from point 3.2 it is necessary to make own transformers.

By analyzing the errors in the columns and the assumptions made, we can see that most of the operations to be performed are repeated. Therefore, it was decided to create a transformer to remove all recurring errors. In addition, a transformer was created to convert a sentence to a number for a Credit_History_Age column and a transformer to replace a string of characters into separate columns for the data contained in the Type_of_Loan column.

```python
# transformer for removing or replacing bad values
    # Possibilities:
    # values with '_' --> Nan
    # outliers (IQR method) --> Nan
    # set min/max values
    # convert datatype
    # replace values

class RepairColumns(BaseEstimator,TransformerMixin):
    def __init__(self, col_name, out_type, min_val=None, max_val=None, no_numerics=False, replace=None, IQR_outliers=False):
        self.col_name = col_name            # column name - string
        self.out_type = out_type            # output column type - string ['Int64'/'float'/'category']
        self.min_val = min_val              # min acceptable value for column - int
        self.max_val = max_val              # max acceptable value for column - int
        self.no_numerics = no_numerics      # check if value is numerci - True/False
        self.replace = replace              # replace 1st value in list for 2nd - list with two elements
        self.IQR_outliers = IQR_outliers    # use outliers filter - True/False
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        # find numbers with bad sign ('_')
        if self.no_numerics:
            X[self.col_name].replace(np.nan, '_', inplace=True, regex=True)
            
            rows_bad_sign = X[X[[self.col_name]].apply(lambda x: not x[0].replace('.', '').isdigit(), axis=1)]
            X.loc[rows_bad_sign.index, self.col_name] = np.nan
        
        # replacing values
        if self.replace: 
            X[self.col_name].replace(self.replace[0], self.replace[1], inplace=True, regex=True)
            
        # change col type
        if self.out_type == 'Int64':
            X[self.col_name] = X[self.col_name].astype(float)
            X[self.col_name] = X[self.col_name].astype(self.out_type)
        else:
            X[self.col_name] = X[self.col_name].astype(self.out_type)
        
        # change bad values to NaN
        if (self.min_val is not None) and (self.max_val is not None):
            rows_bad_values = X[(X[self.col_name] < self.min_val) | (X[self.col_name] > self.max_val)]
            X.loc[rows_bad_values.index, self.col_name] = np.nan
        elif self.min_val is not None:
            rows_bad_values = X[X[self.col_name] < self.min_val]
            X.loc[rows_bad_values.index, self.col_name] = np.nan
        elif self.max_val is not None:
            rows_bad_values = X[X[self.col_name] > self.max_val]
            X.loc[rows_bad_values.index, self.col_name] = np.nan
                             
        # change outliers to NoData value
        if self.IQR_outliers:
            not_null_values = X[X[self.col_name].notnull()][self.col_name]
            q3, q1 = np.percentile(not_null_values, [75 ,25])
            iqr = q3 - q1
            rows_with_outliers = X[(X[self.col_name] < q1 - 1.5 * iqr) | (X[self.col_name] > q3 + 1.5 * iqr)]
            X.loc[rows_with_outliers.index, self.col_name] = np.nan
        
        return X
```

```python
# dedicated transformer for doing OHE on Type_of_Loan column
class RepairTypeOfLoanColumn(BaseEstimator,TransformerMixin):
    def __init__(self, col_name='Type_of_Loan'):
        self.col_name = col_name
        
        # Dict of available type of loans - we assume that there are all types of loan
        self.loan_types = pd.DataFrame({'Loan type': ['Auto Loan', 
                                                      'Credit-Builder Loan', 
                                                      'Personal Loan', 
                                                      'Home Equity Loan', 
                                                      'Not Specified', 
                                                      'Mortgage Loan', 
                                                      'Student Loan', 
                                                      'Debt Consolidation Loan', 
                                                      'Payday Loan', 
                                                      'No Loan']})
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X): 
        # Splitting string with types of loan to separate columns (OHE)
        # Add new columns
        for index, row in self.loan_types.iterrows():
            X[row['Loan type']] = 0

        # Fill in added columns (assign value '1') if specified type of loan occurred in column Type_of_Loan
        loan_types_list = list(self.loan_types['Loan type'])
        for index, row in X.iterrows():
            for loan in loan_types_list:
                try:
                    if loan in row[self.col_name]:
                        X.loc[index, loan] = 1
                except:
                    X.loc[index, 'No Loan'] = 1

        # Remove original column
        X.drop([self.col_name], axis=1, inplace=True)
        
        return X
```

```python
# dedicated transformer for convert sentense into float
class RepairCreditHistoryAgeColumn(BaseEstimator,TransformerMixin):
    def __init__(self, col_name='Credit_History_Age'):
        self.col_name = col_name
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        # replace text for mathematical equation
        X[self.col_name].replace(' Years and ', '+', inplace=True, regex=True)
        X[self.col_name].replace(' Months', '/12', inplace=True, regex=True)

        def string_to_float(row):
            if not type(row[self.col_name]) == float:
                return eval(row[self.col_name])
            else:
                return 999
        X['Credit_History_Age_Numeric'] = X.apply(lambda row: string_to_float(row), axis=1)

        # delete original column
        X.drop([self.col_name], axis=1, inplace=True)

        rows_bad_values = X[X['Credit_History_Age_Numeric'] == 999.0]
        X.loc[rows_bad_values.index, 'Credit_History_Age_Numeric'] = np.nan
        
        return X
```

### Fill in NaN values
<br> 
The data contained a lot of NoData values and additionally any values that did not meet the assumptions or requirements were also replaced with NoData. All such values will be swapped depending on the data type:

- numerical values for median
- categorical values to the most frequent value

```python
# dedicated transformer for fill in NaN values
class FillNa(BaseEstimator,TransformerMixin):
    def __init__(self):
        pass
        
    # function to fill in numeric columns
    def fill_in_numeric(self, df):
        df_numerics = df.select_dtypes(include='number')
        numerics_col_names = df_numerics.columns
        
        SI_numerics = SimpleImputer(missing_values=np.nan, strategy='median') # <-- define strategy
        filed_numerics = SI_numerics.fit_transform(df_numerics)
        filed_numerics = pd.DataFrame(filed_numerics, columns=numerics_col_names)
        return filed_numerics
    
    # function to fill in category columns
    def fill_in_category(self, df):
        df_category = df.select_dtypes(exclude='number')
        category_col_names = df_category.columns
        
        SI_category = SimpleImputer(missing_values=np.nan, strategy='most_frequent') # <-- define strategy
        filed_category = SI_category.fit_transform(df_category)
        filed_category = pd.DataFrame(filed_category, columns=category_col_names)
        return filed_category
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        filed_numerics = self.fill_in_numeric(X)
        filed_category = self.fill_in_category(X)
        
        return pd.concat([filed_numerics, filed_category], axis=1)
```

```python
# dedicated transformer for doing OHE on specific column (using pandas get_dummies)
class OHE(BaseEstimator,TransformerMixin):
    def __init__(self, col_to_OHE):
        self.col_to_OHE = col_to_OHE
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = pd.get_dummies(X, columns=self.col_to_OHE, drop_first=True)
        return X
```

Like most sets for classification, this one is not balanced, so we will try to solve it by removing random records from the excessive class and artificially generating records for the class with the smallest size.

```python
# dedicated transformer for dealing with unbalanced dataset with 3 uniue classes (targets)
class BalanceDataset(BaseEstimator,TransformerMixin):
    def __init__(self, target_col, labels_dict, undersampling=True, oversampling=True, under_factor=0.1, over_factor=0.1):
        self.target_col = target_col               # target column
        self.labels_dict = labels_dict             # classes labels in dict
        self.undersampling = undersampling         # apply undersampling?
        self.oversampling = oversampling           # apply oversampling?
        self.under_factor = under_factor           # the difference between the most numerous and middle class
        self.over_factor = over_factor             # the difference between the least numerous and middle class
        
        # {(freq-key) : (class-value)}
        self.freq_dict = {value:key for (key,value) in dict(pd.DataFrame(self.target_col).value_counts()).items()}
    
    def do_undersampling(self, X, y, undersampling_factor, frequency_dict):
        # list with frequency of classes
        freq_list = list(frequency_dict.keys())
        freq_list.sort()
        
        # convert array to df
        y = pd.DataFrame(y)
        
        amount_to_del = int(freq_list[2] - freq_list[1] * (1 + undersampling_factor))
        class_to_drop = frequency_dict[freq_list[2]][0]
        indices_to_drop = np.random.choice(y[y[0] == class_to_drop].index, amount_to_del, replace=False)
        
        X = pd.DataFrame(X).drop(indices_to_drop)
        y = y.drop(indices_to_drop)
        
        return X, y
    
    def do_oversampling(self, X, y, oversampling_factor, frequency_dict):        
        # list with frequency of classes
        freq_list = list(frequency_dict.keys())
        freq_list.sort()
        
        target_amount = int(freq_list[1] - freq_list[1] * oversampling_factor)
        class_to_overample = frequency_dict[freq_list[0]][0]
        
        oversampling = SMOTE({class_to_overample: target_amount}, random_state=42, n_jobs=-1)
        X_oversample, y_overlample = oversampling.fit_resample(X, y)
        
        return X_oversample, y_overlample
    
    def displ_result(self, y_raw, y_balanced, labels_dict):
        freq_raw = pd.DataFrame(y_raw).value_counts()
        freq_balanced = pd.DataFrame(y_balanced).value_counts()
        
        index = [int(x[0]) for x in list(freq_raw.index)]
        labels = [labels_dict.get(item,item)  for item in index]
        
        plt.rcParams["figure.figsize"] = (15, 5)
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.suptitle('Result of dataset balancing')
        ax1.bar(labels, freq_raw.values)
        ax1.set_title('Frequency before balancing')
        ax2.bar(labels, freq_balanced.values)
        ax2.set_title('Frequency after balancing')
        ax2.axis(ymin=0, ymax=freq_raw.values.max())
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        if self.undersampling == True and self.oversampling == True:
            X_balanced, y_balanced = self.do_undersampling(X, self.target_col, self.under_factor, self.freq_dict)
            X_balanced, y_balanced = self.do_oversampling(X_balanced, y_balanced, self.over_factor, self.freq_dict)
        
        elif self.undersampling == True:
            X_balanced, y_balanced = self.do_undersampling(X, self.target_col, self.under_factor, self.freq_dict)
        
        elif self.oversampling == True:
            X_balanced, y_balanced = self.do_oversampling(X, self.target_col, self.over_factor, self.freq_dict)
        
        self.displ_result(self.target_col, y_balanced, self.labels_dict)
        return X_balanced, y_balanced
```

## Data modeling
Having properly processed data, we can finally start data modeling.

It is worth having a function that will return us several indicators on the basis of which we will evaluate the effectiveness of the models.

```python
def calculate_metrices(model, X_test_dataset, y_test_dataset):
    y_predict = model.predict(X_test_dataset)
    y_pedict_proba = model.predict_proba(X_test_dataset)
    labels = ['Good', 'Poor', 'Stendard']
    
    print(classification_report(y_test_dataset, y_predict, target_names=labels))
    mpl.rcParams.update(mpl.rcParamsDefault)
    ConfusionMatrixDisplay(confusion_matrix(y_test_dataset, y_predict), display_labels = labels).plot()
    plt.show()
    
    f1 = round(f1_score(y_test_dataset, y_predict, average='weighted'), 2)
    recall = round(recall_score(y_test_dataset, y_predict, average='weighted'), 2)
    precision = round(precision_score(y_test_dataset, y_predict, average='weighted'), 2)
    roc_auc = round(roc_auc_score(y_test_dataset, y_pedict_proba, average="weighted", multi_class="ovr"), 2)
    
    return f1, recall, precision, roc_auc
```
