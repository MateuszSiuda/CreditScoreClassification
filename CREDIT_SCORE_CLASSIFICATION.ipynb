## Introduction
<ins>Credit Score</ins> determines the probability that the borrower will pay back the loan on time. <br>

To define client scoring companies use many information which the most important are:
- payment history,
- current unpaid debt,
- the number and type of loan accounts,
- lenght of period of time that loan accounts is open,
- usage of available credit,
- the number of credit inquiries,

This is very important to have a good credit score because it can help us get a loan for a new IPhone, car, house but also get a new policy or rent an apartment.

**In this project we will try to classify client Credit score into three buckets: Poor, Standard, Good, having available person's credit-related information**

**Author's comment:** <br>

We can see three types of error in column Age:
- negative values, eg. -500,
- outliers, eg. 7518,
- values with underscore, eg. 33_

We can notice that many records refer to one client (company is checking credit score every month before prolonging credit limit) and every time at least one record has appropriate value so we can replace wrong value to good one.

```python
# Function to repair Age column
def repair_age_column(df):
    # replace wrong value to good one from another record if it exist for that client
    for index, row in df.iterrows():
        if not row['Age'].isdigit() or int(row['Age']) < 0 or int(row['Age']) > 70:
            customer_age_list = list(df[df['Customer_ID'] == row['Customer_ID']]["Age"])
            customer_age_list = [int(x) for x in customer_age_list if x.isdigit() and int(x) > 0 and int(x) < 70]
            # choose the middle value
            age_index = int(np.round(len(customer_age_list)/2 + 0.1, decimals=0)) - 1
            if age_index >= 0:
                df.loc[index, 'Age'] = str(customer_age_list[age_index])
            else:
                continue  
        
    # convert data type
    df[['Age']] = df[['Age']].astype(int)
```

'Occupation' column contain 15 unique occupations and something what we can call NoData (_______).

Missing values will be filled on the base of the other rows. In case when we won't find information about client occupation we will assign 'Other' value.

```python
# Function to repair Occupation column
def repair_occupation_column(df):
    # replace wrong value if it possible
    for index, row in df.iterrows():
        if row['Occupation'] == '_______':
            customer_occupation_list = list(df[df['Customer_ID'] == row['Customer_ID']]["Occupation"])
            customer_occupation_list = [x for x in customer_occupation_list if x != '_______']
            if len(customer_occupation_list) >= 0:
                df.loc[index, 'Occupation'] = customer_occupation_list[0]
            else:
                continue  
    
    df["Occupation"].replace('_______', "Other", inplace=True, regex=True)
        
    # convert column type
    df[['Occupation']] = df[['Occupation']].astype('category')
```

In this column ocuure two types of error which should be removed:
- _ sign, eg. 35547.71_

```python
# Function to repair Annual income column
def repair_annual_income_column(df):
    # remove bad sign
    df["Annual_Income"].replace('_', '', inplace=True, regex=True)
        
    # convert column type
    df[['Annual_Income']] = df[['Annual_Income']].astype('float')
    
    # We will chceck if there is no error for value >200k$
    for index, row in df.iterrows():
        if row['Annual_Income'] > 200000 or row['Annual_Income'] < 0:
            q3, q1 = np.percentile(df[df['Customer_ID'] == row['Customer_ID']]["Annual_Income"], [75 ,25])
            iqr = q3 - q1
            if (row['Annual_Income'] < q1 - 1.5 * iqr) or (row['Annual_Income'] > q3 + 1.5 * iqr):
                median = np.median(df[df['Customer_ID'] == row['Customer_ID']]["Annual_Income"])
                df.loc[index, 'Annual_Income'] = median
            
            # Additionally we will check really big annual income (>1kk$) 
            if row['Annual_Income'] > 1000000:
                mode = stats.mode(df[df['Customer_ID'] == row['Customer_ID']]["Annual_Income"])[0][0]
                df.loc[index, 'Annual_Income'] = mode
                
```

**Author's comment:**<br>

We have to split this column for new ones that will be indicate if specified type of credit occurred (One Hot Encoding)

```python
# Function to repair Type of Loan column
def repair_type_of_loan_column(df):
    # Finding unique values for the type of loan. 'None' value is for customers with no loan
    loan_types = [x for x in list(df.Type_of_Loan.unique()) if not type(x) == float]
    loan_types = ', '.join(loan_types).replace('and ', '')
    loan_types = pd.DataFrame(loan_types.split(', '))
    loan_types.drop_duplicates(inplace=True)
    loan_types.reset_index(drop=True, inplace=True)
    loan_types.set_axis(['Loan type'], axis=1, inplace=True)
    loan_types = loan_types.append({'Loan type': 'No Loan'}, ignore_index=True)
    
    # Add new columns
    for index, row in loan_types.iterrows():
        df[row['Loan type']] = 0
        
    # Fill in added columns (assign value '1') if specified type of loan occurred in column Type_of_Loan
    loan_types_list = list(loan_types['Loan type'])
    for index, row in df.iterrows():
        for loan in loan_types_list:
            try:
                if loan in row['Type_of_Loan']:
                    df.loc[index, loan] = 1
            except:
                df.loc[index, loan] = 1
    
    # Remove original column
    df.drop(['Type_of_Loan'], axis=1, inplace=True)
```

```python
# function to repair Credit Mix column
def repair_credit_mix_column(df):
    credit_dict = {'Good': 1,
                  'Standard': 2,
                  'Bad': 3,
                  '_': 4}

    credit_dict_reverse = {1: 'Good',
                           2: 'Standard',
                           3: 'Bad',
                           4: '_'}

    # replace '_' sign for most frequent occured value
    for index, row in df.iterrows():
        if row['Credit_Mix'] == '_':
            client_credit_mix = df[df['Customer_ID'] == row['Customer_ID']]["Credit_Mix"]          
            client_credit_mix = client_credit_mix.replace(credit_dict)
            client_credit_mix.drop(client_credit_mix.index[client_credit_mix == 4], inplace=True)

            mode = stats.mode(client_credit_mix)[0][0]
            df.loc[index, 'Credit_Mix'] = credit_dict_reverse[mode]
            
    df.Credit_Mix = df.Credit_Mix.astype('category')
```

```python
def feature_enginnering(df, OHE=True, discret_int=False, discret_float=False, standarize=True, PCA=False):
    
    # Split data to X and y
    X_transform = df.loc[:, df.columns != 'Credit_Score']
    y = pd.DataFrame(df.Credit_Score).loc[:, ['Credit_Score']].apply(LabelEncoder().fit_transform)
    
    # applay One Hot Encoder on categorical data
    if OHE == True:
        columns_to_OHE = ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']
        X_transform = pd.get_dummies(X_transform, columns=columns_to_OHE, drop_first=True)
    
    # Discretize int data. Bin continuous data into 4 intervals
    if discret_int == True:
        int_col_to_descretize = ['Age', 
                                 'Num_Bank_Accounts', 
                                 'Num_Credit_Card', 
                                 'Num_of_Loan', 
                                 'Num_of_Delayed_Payment', 
                                 'Num_Credit_Inquiries', 
                                 'Changed_Credit_Limit', 
                                 'Credit_History_Age_Numeric', 
                                 'Delay_from_due_date']

        KBD_int = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='quantile')
        X_transform[int_col_to_descretize] = KBD_int.fit_transform(X_transform[int_col_to_descretize])
    
    # Discretize float data. Bin continuous data into 10 intervals
    if discret_float == True:
        float_col_to_descretize = ['Monthly_Inhand_Salary', 
                                   'Interest_Rate', 
                                   'Outstanding_Debt', 
                                   'Credit_Utilization_Ratio', 
                                   'Total_EMI_per_month', 
                                   'Amount_invested_monthly', 
                                   'Monthly_Balance']

        KBD_float = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='quantile')
        X_transform[float_col_to_descretize] = KBD_float.fit_transform(X_transform[float_col_to_descretize])
    
    # Standaize data
    if standarize == True:
        X_transform = StandardScaler().fit_transform(X_transform)
    
    # PCA
    if PCA == True:
        pca = PCA(n_components=40)
        pca.fit(X_transform)
        print(sum(pca.explained_variance_ratio_))
        X_transform = pca.transform(X_transform)
    
    X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(X_transform), y, test_size=0.2, random_state=42)
    
    return X_train, X_test, y_train, y_test
```

```python
def calculate_metrices(model, X_test_dataset, y_test_dataset):
    y_predict = model.predict(X_test_dataset)
    y_pedict_proba = model.predict_proba(X_test_dataset)
    labels = ['Good', 'Poor', 'Stendard']
    
    print(classification_report(y_test_dataset, y_predict, target_names=labels))
    mpl.rcParams.update(mpl.rcParamsDefault)
    ConfusionMatrixDisplay(confusion_matrix(y_test_dataset, y_predict), display_labels = labels).plot()
    plt.show()
    
    f1 = round(f1_score(y_test_dataset, y_predict, average='weighted'), 2)
    recall = round(recall_score(y_test_dataset, y_predict, average='weighted'), 2)
    precision = round(precision_score(y_test_dataset, y_predict, average='weighted'), 2)
    roc_auc = round(roc_auc_score(y_test_dataset, y_pedict_proba, average="weighted", multi_class="ovr"), 2)
    
    return f1, recall, precision, roc_auc
    
```

```python
# define models, paramiters, statistics
models = ['LogisticRegression', 
          'SVC', 
          'RandomForestClassifier', 
          'AdaBoostClassifier', 
          'XGBClassifier', 
          'GaussianNB', 
          'KNeighbors']

paramiters = [{LogisticRegression(): {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                                      'max_iter': [10, 30, 50, 100, 200]}},
         
              {SVC() : {SVC() : {'C': [0.0001, 0.001, 0.01, 0.1, 1],
                                 'degree': [2, 3, 5, 7, 10],
                                 'probability': [True]}},
              
              {RandomForestClassifier(): {'n_estimators': [20, 50, 100, 150, 200, 300, 500],
                                          'min_samples_leaf': [1, 2, 3, 5],
                                          'criterion': ['gini', 'entropy'],
                                          'max_depth': [5, 10, 30, 50]}},
              
              {AdaBoostClassifier(): {'n_estimators': [30, 50, 100, 300, 500, 700],
                                     'learning_rate': [0.1, 1, 10]}},
              
              {XGBClassifier(): {'use_label_encoder': [False],
                                 'learning_rate': [0.1, 0.5, 1],
                                 'min_split_loss': [0.00001, 0.0001, 0.001, 0.01, 0.1],
                                 'max_depth': [50, 100, 150]}},
              
              {GaussianNB(): {}},
              
              {KNeighborsClassifier(): {'n_neighbors': [3, 4, 5, 7],
                                        'leaf_size': [20, 30, 50, 100]}}
             ]

models_stats = pd.DataFrame({'model': [], 
                             'f1_score': [], 
                             'recall_score': [], 
                             'precision_score': [], 
                             'roc_auc_score': []})
```
